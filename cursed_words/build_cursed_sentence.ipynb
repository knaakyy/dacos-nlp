# unsmile 악플/욕설 추출

두 TSV(`unsmile_train_v1.0.tsv`, `unsmile_valid_v1.0.tsv`)을 합쳐 `악플/욕설` 값이 1인 행만 필터링해 `cursed_sentence.csv`로 저장합니다.

실행: 이 노트북 셀을 순서대로 실행하거나, 스크립트가 필요하면 `build_cursed_sentence.py`와 동일한 로직을 사용하세요.

```python
from pathlib import Path

import pandas as pd

base_dir = Path.cwd()
train_path = base_dir / "cursed_words/unsmile_train_v1.0.tsv"
valid_path = base_dir / "cursed_words/unsmile_valid_v1.0.tsv"
out_path = base_dir / "cursed_words/cursed_sentence.csv"

train_df = pd.read_csv(train_path, sep="\t")
valid_df = pd.read_csv(valid_path, sep="\t")

combined = pd.concat([train_df, valid_df], ignore_index=True)
filtered = combined[combined["악플/욕설"] == 1]

# Excel 호환을 위해 UTF-8 with BOM으로 저장 (Excel에서 깨짐 방지)
filtered.to_csv(out_path, index=False, encoding="utf-8-sig")
print(f"saved {len(filtered)} rows -> {out_path.relative_to(base_dir)}")
```
